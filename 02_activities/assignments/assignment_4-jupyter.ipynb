```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data loading code)
data = {
    'Timestamp': ['2013-04-10 12:10:39', '2013-04-10 12:30:55', '2013-04-10 12:33:49', '2013-04-10 12:52:55', '2013-04-10 13:24:47'],
    'How would you describe your level of physical health?': ['Good', 'Excellent', 'Good', 'Good', 'Very good']
}

# Create DataFrame
df = pd.DataFrame(data)

# Count occurrences of each unique value in the column
health_counts = df['How would you describe your level of physical health?'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
health_counts.plot(kind='bar', color='skyblue')

# Adding titles and labels
plt.title('Distribution of Physical Health Descriptions', fontsize=16)
plt.xlabel('Physical Health Description', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
